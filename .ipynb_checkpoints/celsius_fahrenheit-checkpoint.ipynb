```python
import pandas as pd
import seaborn as sb

datos = pd.read_csv("Celsiust.csv")
```

```python
datos.info()
```

```python
datos.head()
```

```python
sb.scatterplot(x="Celsius", y="fahrenheit", data=datos, hue="fahrenheit", palette="coolwarm")
```

```python
X = datos["Celsius"]
y = datos["fahrenheit"]
```

```python
X
```

```python
y
```

```python
process_X= X.values.reshape(-1,1)
process_y= y.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression 

modelo= LinearRegression()
```

```python
modelo.fit(process_X, process_y)
```

```python
modelo.predict([[7900]])
```

```python
modelo.score(process_X, process_y)
```
